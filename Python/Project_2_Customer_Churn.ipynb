```python
import pandas as pd
import os

# Load dataset
file_path = r"C:\Users\SKY\Desktop\JOB prep\Project-2 Customer Churn\Data\WA_Fn-UseC_-Telco-Customer-Churn.csv"
df = pd.read_csv(file_path)

print("Initial shape:", df.shape)
df.head()
```

```python
# Check duplicates
print("Duplicate customerIDs:", df['customerID'].duplicated().sum())

# Check missing values
print("\nMissing values per column:\n", df.isnull().sum())

# Data types overview
print("\nData types:\n", df.dtypes)
```

```python
# Convert TotalCharges to numeric (coerce errors -> NaN)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Fill blanks with 0
df['TotalCharges'] = df['TotalCharges'].fillna(0)

print("TotalCharges conversion done. Nulls left:", df['TotalCharges'].isnull().sum())
```

```python
# Mapping for binary conversion
binary_map = {"Yes": 1, "No": 0, "Male": 1, "Female": 0}

# Identify all columns with only Yes/No or Male/Female
binary_candidate_cols = []
for col in df.columns:
    unique_vals = set(df[col].unique())
    if unique_vals.issubset({"Yes", "No"}):
        binary_candidate_cols.append(col)
    if unique_vals.issubset({"Male", "Female"}):
        binary_candidate_cols.append(col)

print("Binary candidate columns:", binary_candidate_cols)
```

```python
# Create binary versions of identified columns
for col in binary_candidate_cols:
    df[col + "_bin"] = df[col].map(binary_map)

print("✅ Binary columns created:")
print(df.filter(like="_bin").head())

```

```python
# Save cleaned dataset in same folder with _cleaned suffix
cleaned_path = os.path.splitext(file_path)[0] + "_cleaned.csv"
df.to_csv(cleaned_path, index=False)

print(f"✅ Cleaned dataset saved at: {cleaned_path}")
print("Final shape:", df.shape)
```
